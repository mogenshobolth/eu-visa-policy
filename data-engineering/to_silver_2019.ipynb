```python
import pandas as pd
import numpy as np
df = pd.read_excel('../data/bronze/2019-consulates-schengen-visa-stats.xlsx',1)
```

```python
df = df.rename(columns = {"Schengen State": "eu_state"
                         ,"Country where consulate is located": "origin_country"
                         ,"Consulate": "origin_consulate"
                         ,"Uniform visas applied for": "visas_applied"
                         ,"Total  uniform visas issued (including MEV) \n": "visas_issued"
                         ,"Uniform visas not issued": "visas_not_issued"})
df = df[["eu_state", "origin_country","origin_consulate","visas_applied", "visas_issued", "visas_not_issued"]]
```

```python
df[df["eu_state"].isna()]
```

```python
df = df[df["eu_state"].notna()]
```

```python
df["visas_applied"].replace(np.nan, 0, inplace=True)
df["visas_issued"].replace(np.nan, 0, inplace=True)
df["visas_not_issued"].replace(np.nan, 0, inplace=True)
df["visa_refusal_rate"] = df["visas_not_issued"] / (df["visas_issued"] + df["visas_not_issued"])
```

```python
df[df["visas_applied"] != df["visas_issued"] + df["visas_not_issued"]]
```

```python
df[df["visa_refusal_rate"]>1]
```

```python
year = 2019
df["year"] = year
```

```python
df.to_csv('../data/silver/schengen-visa-' + str(year) + '.csv')
```
